```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("advertising.csv")
x=dataset[['TV']]
y=dataset['Sales']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=4)

pol=PolynomialFeatures(degree=3)
x_train_poly=pol.fit_transform(x_train)
x_test_poly=pol.transform(x_test)

```

```python
model=LinearRegression()
model.fit(x_train_poly,y_train)
y_pred=model.predict(x_test_poly)

print(f"mean squared error{mean_squared_error(y_test,y_pred)}")
print(f"mean absolute error{mean_absolute_error(y_test,y_pred)}")
print(f"r2 score{r2_score(y_test,y_pred)}")
```

```python
plt.figure(figsize=(10,6))

x_range=np.linspace(x_test.min(),x_test.max(),100)

x_range_poly=pol.transform(x_range)
y_poly=model.predict(x_range_poly)



plt.scatter(x_test,y_test,c='red',label="predicted values")

plt.plot(x_range,y_poly,c='blue',lw=2,label="polynomial curve")
plt.xlabel("predicted")
plt.ylabel("actual")
plt.title('predicted vs actual')
plt.legend()
plt.show()
```
